```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kmeans_scratch(X, k, max_iters=100):

  np.random.seed(16)
  idx = np.random.choice(len(X), k , replace=False)
  centroids = X[idx]

  for _ in range(max_iters):
    dis = np.array([np.sqrt(((X - centroid) ** 2).sum(axis=1))
                            for centroid in centroids])
    labels = np.argmin(dis, axis=0)
    new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

    if np.all(centroids == new_centroids):
      break
    centroids = new_centroids
  return labels, centroids
```

```python
np.random.seed(16)
points = np.vstack([
    np.random.normal(0, 0.5, (50, 2)),
    np.random.normal(3, 0.5, (50, 2)),
    np.random.normal(-2, 0.5, (50, 2))])
```

```python
k = 7
labels, centroids = kmeans_scratch(points, k)
```

```python
plt.scatter(points[:, 0], points[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:,0], centroids[:, 1], c = 'red', marker='x', s=200, linewidths=3)
plt.title("K-means clustering from scratch")
plt.show()
```
